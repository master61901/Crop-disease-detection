```python
!pip install tensorflow pillow
!pip install tensorflow
import tensorflow as tf
import tkinter as tk
from tkinter import filedialog
from PIL import ImageTk, Image
import numpy as np
from tensorflow.keras.models import load_model

# Load the model
model = load_model(r'C:\Users\bahug\OneDrive\Desktop\crop disease\model.h5')
model.summary()

# Preprocess image function
def preprocess_image(image_path):
    img = Image.open(image_path)
    img = img.resize((100,100))  # Adjust based on model
    img_array = np.array(img) / 255.0  # Normalize if needed
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    return img_array

# Predict function
def predict_image():
    file_path = filedialog.askopenfilename()
    if file_path:
        # Display the image
        img = Image.open(file_path)
        img = img.resize((200, 200))  # Resize for display
        img = ImageTk.PhotoImage(img)
        label_image.configure(image=img)
        label_image.image = img

        # Preprocess and predict
        input_data = preprocess_image(file_path)
        prediction = model.predict(input_data)
        label_result.configure(text=f"Prediction: {np.argmax(prediction)}")  # Adjust output format

# GUI setup
window = tk.Tk()
window.title("Image Classifier")

label_image = tk.Label(window)
label_image.pack()

button_select = tk.Button(window, text="Select Image", command=predict_image)
button_select.pack()

label_result = tk.Label(window, text="Prediction: ")
label_result.pack()

window.mainloop()
```
